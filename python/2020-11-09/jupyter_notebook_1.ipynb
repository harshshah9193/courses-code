```python
# data on 5 bodies
# masses between 0-1
# velocities between 0-10

np.random.seed(1)

masses = np.random.random(5)
velocities = np.random.random(5) * 10
velocities

# kinetic energy

energies = 1/2 * masses * velocities**2
print(energies)
print(velocities)

velocities

total_energy = np.sum(energies)
print(total_energy)
```

```python
# filter out any values below 0.01
valid_value_indicators = energies >= 0.01
energies_filtered = energies[valid_value_indicators]
print(energies_filtered)

# replace values below 0.01 with NaN
invalid_value_indicators = energies < 0.01

energies_with_nan = energies.copy()
energies_with_nan[invalid_value_indicators] = np.nan
print(energies_with_nan)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2*np.pi, 0.01)
sin_x = np.sin(x)
cos_x = -np.cos(x)

plt.plot(x, sin_x, label='sin(x)')
plt.plot(x, cos_x, label="cos(x)")

plt.title("trigonometric functions")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.legend()
plt.axis("equal")
```

```python
# optional
def gaussian_data(mu, sigma):
    x = np.arange(mu-2*sigma, mu+2*sigma, 0.01)
    y = x*sigma
    return (x, y)

x, y = gaussian_data(1, 1)
# optional

def draw_gaussian_curve(mu, sigma, text):
    x = np.arange(mu-3*sigma, mu+3*sigma, 0.01)
    y = 1/(sigma*np.sqrt(2*np.pi)) * np.exp(
        (-(x-mu)**2) / (2*sigma**2)
    )

    color = (1, 0, 0)

    plt.plot(x, y, color=color, marker=",", linestyle="none", label=text)

draw_gaussian_curve(0, 1, "gauss 1")
draw_gaussian_curve(0.5, 2, "gauss 2")
draw_gaussian_curve(1, 0.5, "gauss 3")
plt.legend()
```
